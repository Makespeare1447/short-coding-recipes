```python
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

# activate custom stylesheet:
matplotlib.style.use('signature.mplstyle')

# set palette:
#n_colors = 8
#cmap = sns.color_palette("husl", n_colors)

cmap = [
    '#08F7FE',  
    '#FE53BB',  
    '#F5D300',  
    '#00ff41'
]
sns.palplot(cmap)
```

```python
# some synthetic data:
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
fillrange = [3,5]

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
ax.set_prop_cycle(color=cmap)
ax.set_title('demonstration plot')

ax.plot(x, y, label='sin(x)', alpha = 0.6, marker='o')
ax.plot(x, y2, label='cos(x)', alpha = 0.6, marker='o')
ax.fill_between(x, y, y2, where=x>5 , label='fill area', alpha = 0.4, color=cmap[2])

# set custom rotation for x-axis labels:
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    
# annotate plot:
ax.annotate(text='annotation string', xy=(0.3, 0.88), xycoords='axes fraction', size=15, bbox=dict(facecolor='white', edgecolor='gray', linewidth=2))

legend = ax.legend()
```
